```python
import requests as re
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://www.webometrics.info/en/europe/turkey" # We define the link to a variable named url.

data = re.get(url).text
#Then we write our url variable into it with the get() function using the requests library. Then we convert it to a font with the .text property.

soup = BeautifulSoup(data,'html.parser')
#Here we review parts of it with the BeautifulSoup library.

for table in soup.find_all('table'):# We find the tables in the data source by creating a for loop.
    print(table.get('class'))
    
```

```python
# We search for the name of the table we want to find. 
table = soup.find('table',class_ ='sticky-enabled')

#We create a dataframe and enter the names of the columns. 
df = pd.DataFrame(columns = ['Rank','World Rank','University','Impact Rank','Openness Rank','Excellence Rank'])

# Then we create a for loop so that we can get the rows in the table one by one. 
for row in table.tbody.find_all('tr'):
    columns = row.find_all('td')
    
    if(columns != []):
        rank = columns[0].text.strip()
        world_rank = columns[1].text.strip()
        university = columns[2].text.strip()
        ımpact_rank = columns[4].text.strip()
        open_rank = columns[5].text.strip()
        excellence_rank = columns[6].text.strip()
        
        # Then these values are added to the DataFrame we created. 
        df = df.append({'Rank':rank,'World Rank':world_rank,'University':university,'Impact Rank':ımpact_rank,'Openness Rank':open_rank,'Excellence Rank':excellence_rank},
                       ignore_index=True)
df.head()
```
